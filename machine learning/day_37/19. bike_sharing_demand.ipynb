#### 2011년 1월 ~ 2012년 1월
년,월,일,시간
* 날짜,/시간, 기온, 습도, 풍속 등의 정보를 기반으로 1시간 간격 동안의 자전거 대여 횟수가 기재되어 있음

datetime - hourly date + timestamp <br>
season -  1 = spring, 2 = summer, 3 = fall, 4 = winter <br>
holiday - 1=토, 일주일의 주말을 제외한 국경일 등의 휴일, 0=휴일이 아닌 날 <br>
workingday - 1=토, 일주일의 주말 및 휴일이 아닌 주중, 0 = 주말 및 휴일 <br>
weather - <br>
1: 맑음, 약간 구름 낀 흐림 <br>
2: 안개, 안개+흐림 <br>
3: 가벼운 눈, 가벼운 비 + 천둥 <br>
4: 심한 눈/비, 천둥/번개 <br>
temp - 온도(섭씨) <br>
atemp - 체감온도(섭씨) <br>
humidity - 상대습도 <br>
windspeed - 풍속 <br>
casual - 사전에 등록되지 않는 사용자가 대여한 횟수 <br>
registered - 사전에 등록된 사용자가 대여한 횟수 <br>
count - 대여 횟수 <br>
