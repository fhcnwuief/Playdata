```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
path = '../../driver/chromedriver_win32/chromedriver.exe'
driver = webdriver.Chrome(service=Service(path))
import time
driver.get('http://kopia.or.kr/info/statistics.php')
time.sleep(3)
```

### 사이트 접속

```python
from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
content = soup.select('div.scroll_cont')
content
```

```python
import pandas as pd
content_html = str(content)

content_df_list = pd.read_html(content_html)
content_df = content_df_list[0]
content_df.excel=('../../output/content.xlsx')
```
