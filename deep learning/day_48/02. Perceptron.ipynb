### 2.1 퍼셉트론이란?
- 퍼셉트론(인공뉴런): 다수의 신호를 입력으로 받아 하나의 신호를 출력
    - 신호: 흐른다(1), 흐르지 않는다(0) 두 가지 값을 가짐.

입력신호: x1, x2
출력신호: y
가중치: w1, w2(w는 weight의 머리글자)

- 퍼셉트론에서는 가중치가 클수록 강한 신호를 흘림.
- 전류에서 말하는 저항. 저항이 낮을 수록 큰 전류가 흐름.

뉴런, 노드: x1, x2, y
입력신호가 뉴런에 보내질 때 가각 고유한 가중치가 곱해짐. (w1x1, w2x2)

- 뉴런이 활성화: 뉴런에서 보내온 신호의 총합이 정해진 한계를 넘어설 때만 1을 출력
- 임계값: 정해진 한계. 세타 기호로 표현

###### 1. AND 게이트

퍼셉트론으로 표현하려면 진리표대로 작동하는 w1, w1, 세타 값을 정해야 함

(w1, w2, 세타) = (0.5, 0.5, 0.7), (0.5, 0.5, 0.8) 또는 (1.0, 1.0, 1.0)

x1, x2가 모두 1일 때만 가중신호의 총합이 주어진 임계값을 넘게 됨.

- AND 진리표





```python
def AND(x1,x2):
    w1,w2, theta = 0.5, 0.5, 0.7
    # 초기 값인 가중치1 가중치2 임계값 설정
    tmp = w1*x1 + w2*x2
    if tmp <= theta: # theta보다 작으면 return 0
        return 0
    elif tmp > theta: # theta보다 크면 return 1
        return 1
    
# (w1,w2,θ) = 0.5 0.5 0.7 이라 하면
# x1=0, x2=0 일때 0.5"*0 + 0.5*0 = 0 < 0.7 이므로 0 출력
# x1=0, x2=1 일때 0.5"*0 + 0.5*1 = 0.5 < 0.7 이므로 0 출력
# x1=1, x2=0 일때 0.5"*1 + 0.5*0 = 0.5 < 0.7 이므로 0 출력
# x1=1, x2=1 일때 0.5"*1 + 0.5*1 = 1 > 0.7 이므로 1 출력
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

###### 2. NAND게이트(NOT AND)
NAND = not AND. AND 게이트의 출력을 뒤집은 것

- NAND 게이트 진리표



AND 게이트를 구현하는 매개변수의 부호를 모두 반전하면 NAND 게이트가 됨

```python
def NAND(x1,x2):
    w1,w2, theta = -0.5, -0.5, -0.7
    # 초기 값인 가중치1 가중치2 임계값 설정
    tmp = w1*x1 + w2*x2
    if tmp <= theta: # theta보다 작으면 return 0
        return 0
    elif tmp > theta: # theta보다 크면 return 1
        return 1
    
# (w1,w2,θ) = -0.5 -0.5 -0.7 이라 하면
# x1=0, x2=0 일때 -0.5"*0 + -0.5*0 = 0 > -0.7 이므로 1 출력
# x1=0, x2=1 일때 -0.5"*0 + -0.5*1 = -0.5 > -0.7 이므로 1 출력
# x1=1, x2=0 일때 -0.5"*1 + -0.5*0 = -0.5 > -0.7 이므로 1 출력
# x1=1, x2=1 일때 -0.5"*1 + -0.5*1 = -1 < 0.7 이므로 0 출력
```

```python
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
```

3. OR게이트

- OR 게이트 진리표



```python
def OR(x1,x2):
    w1,w2, theta = 0.5, 0.5, 0.3
    # 초기 값인 가중치1 가중치2 임계값 설정
    tmp = w1*x1 + w2*x2
    if tmp <= theta: # theta보다 작으면 return 0
        return 0
    elif tmp > theta: # theta보다 크면 return 1
        return 1
    
# (w1,w2,θ) = 0.5 0.5 0.3 이라 하면
# x1=0, x2=0 일때 0.5"*0 + 0.5*0 = 0 < 0.3 이므로 0 출력
# x1=0, x2=1 일때 0.5"*0 + 0.5*1 = 0.5 > 0.3 이므로 1 출력
# x1=1, x2=0 일때 0.5"*1 + 0.5*0 = 0.5 > 0.3 이므로 1 출력
# x1=1, x2=1 일때 0.5"*1 + 0.5*1 = 1 > 0.3 이므로 1 출력
```

```python
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

### 2-3 가중치와 편향 구현하기

가중치(w1, w2)는 각 입력 신호가 결과에 주는 영향력(중요도)을 조절하는 매개변수

편향(b)은 뉴런이 얼마나 쉽게 활성화(결과로 1을 출력)하느냐를 조정하는 매개변수

예시) b가 -0.1이면 가중치를 곱한 합이 0.1을 초과할 때 뉴런이 활성화.

편향의 의미: 한 쪽으로 치우쳐 균형을 깬다.

```python
# 가중치와 편향을 도입한 AND 게이트 구현

def AND(x1, x2):
    x = np.array([x1, x2]) # x1, x2 배열 생성
    w = np.array([0.5, 0.5])
    b = -0.7 # 편향에 -0.7
    tmp = np.sum(w*x) + b # 활성화 식
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
# 가중치와 편향을 도입한 NAND 게이트 구현
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) # AND와는 가중치(w와 b)만 다르다.
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
# 가중치와 편향을 도입한 OR 게이트 구현
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) # AND와는 가중치(w와 b)만 다르다.
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```python
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
```

```python
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y

print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```
